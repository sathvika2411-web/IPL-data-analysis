```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


plt.rcParams['font.family'] = 'Segoe UI Emoji'

deliveries_data = {
    'batsman': ['Kohli', 'Dhoni', 'Kohli', 'Rohit', 'Kohli', 'Rohit', 'Dhoni'],
    'batsman_runs': [10, 15, 25, 30, 10, 20, 5],
}
deliveries = pd.DataFrame(deliveries_data)


top_batsmen = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)


plt.figure(figsize=(8,5))
sns.barplot(x=top_batsmen.values, y=top_batsmen.index, hue=top_batsmen.index, palette="viridis", legend=False)
plt.title('Top Run Scorers')
plt.xlabel('Runs')
plt.ylabel('Batsman')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


deliveries_data = {
    'bowler': ['Bumrah', 'Ashwin', 'Bumrah', 'Chahal', 'Chahal', 'Ashwin', 'Bumrah'],
    'player_dismissed': ['Dhoni', None, None, 'Rohit', None, None, 'Kohli'],
    'dismissal_kind': ['caught', None, None, 'bowled', None, None, 'lbw']
}
deliveries = pd.DataFrame(deliveries_data)


valid_wickets = deliveries[deliveries['dismissal_kind'].notna()]
top_bowlers = valid_wickets.groupby('bowler')['player_dismissed'].count().sort_values()


plt.figure(figsize=(8,5))
top_bowlers.plot(kind='barh', color='RED')
plt.title('ðŸŽ¯ Top Wicket Takers')
plt.xlabel('Wickets')
plt.ylabel('Bowler')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


matches_data = {
    'match_id': [1, 2],
    'team1': ['RCB', 'MI'],
    'team2': ['CSK', 'RCB'],
    'winner': ['RCB', 'RCB']
}
matches = pd.DataFrame(matches_data)


match_wins = matches['winner'].value_counts()


plt.figure(figsize=(6,6))
match_wins.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=["orange"])
plt.title('ðŸ† Match Wins Percentage')
plt.ylabel('')
plt.show()
```
